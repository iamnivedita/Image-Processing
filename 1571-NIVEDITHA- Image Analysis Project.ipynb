### Import necessary libraries

```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt
```

### Open 'image2.jpg' from the data folder using open cv and display it in the notebook. Make sure to correct for the RGB order

```python
img = cv2.imread('image2.jpg')
print(img)
```

##### Before RGB correction

```python
plt.imshow(img)
```

### After correction

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Draw an empty RED rectangle around toms face and display the image in the notebook

```python
cv2.rectangle(img_rgb, pt1 = (250,400), pt2 = (600,750), color = (255,0,0), thickness = 5)
plt.imshow(img_rgb)
```

### Draw a text named 'TOM' above toms face

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_rgb, text = 'TOM', org = (320,250), fontFace = font, fontScale = 4, color = (0,0,0), thickness = 10, lineType = cv2.LINE_AA)
plt.imshow(img_rgb)
```

### Do the same steps for 'Jerry'

```python
cv2.rectangle(img_rgb, pt1 = (100,400), pt2 = (350,700), color = (255,0,0), thickness = 5)
plt.imshow(img_rgb)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img_rgb, text = 'JERRY', org = (10,390), fontFace = font, fontScale = 4, color = (0,0,0), thickness = 10, lineType = cv2.LINE_AA)
plt.imshow(img_rgb)
```

```python
cv2.imwrite('new_picture.jpg', img_rgb)
```
